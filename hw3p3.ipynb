## Classfier Comparison

```python
names = [
    "Naive Bayes",
    "Logistic Regression",
    "QDA",
    "SVM RBF",
    "Decision Tree",
    "KNN (k=1)"
]

classifiers = [
    GaussianNB(),
    LogisticRegression(random_state=seed),
    QuadraticDiscriminantAnalysis(),
    SVC(kernel='rbf', random_state=seed),
    DecisionTreeClassifier(max_depth=7, random_state=seed),
    KNeighborsClassifier(1),

]
```

```python
def plot_decision_boundaries(X, y, clf, name, dataset_name, ax):

    # Train and Test Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)
    # Fitting the classifier model
    clf.fit(X_train, y_train)
    # Accuracy Train
    y_pred_train = clf.predict(X_train)
    accuracy_train = accuracy_score(y_train, y_pred_train)
    # Accuracy Test
    y_pred = clf.predict(X_test)
    accuracy_test = accuracy_score(y_test, y_pred)

    # Plot decision boundary for each classifier 
    h = 0.02 
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.PiYG)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.PiYG)
    ax.set_title(f"{name}\nTest Accuracy: {accuracy_test:.2f}\nTrain Accuracy: {accuracy_train:.2f}")

fig, axs = plt.subplots(len(datasets), len(classifiers), figsize=(25, 15))
plt.style.use('ggplot')

for i, (X, y, dataset_name) in enumerate(datasets):
    for j, (name, clf) in enumerate(zip(names, classifiers)):
        plot_decision_boundaries(X, y, clf, name, dataset_name, axs[i, j])

plt.tight_layout()
plt.show()
```
